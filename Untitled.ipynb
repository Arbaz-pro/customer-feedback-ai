```python
import re
def split_into_parts(comment: str):
    parts = re.split(r'\bbut\b|\band\b|\bhowever\b|\balthough\b|,|\.', comment)
    print(parts)
    return [p.strip() for p in parts if p.strip()]
```

```python
a="i har ordered it 2 days ago, the product is good but the packaging was damaged and the website is very sow but the delivery was fast"
```

```python
b=split_into_parts(a)
```
